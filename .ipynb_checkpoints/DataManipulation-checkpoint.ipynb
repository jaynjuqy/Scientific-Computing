```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading and inspecting a .csv file
# Uses the libraries Pandas, matplotlib and seaborn for viaualization and manipulation of this data

```python
data = pd.read_csv("./ibtracs.ACTIVE.list.v04r01.csv", na_values=['?', 'N/A', '', 'unknown'])
data.head()
```

```python
data.drop(index=0, inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
print(data.columns)
```

```python
data['STORM_SPEED']=pd.to_numeric(data['STORM_SPEED'], errors='coerce')
data['STORM_DIR']=pd.to_numeric(data['STORM_DIR'], errors='coerce')
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

## Statistical methods covering mean, mode, median, std deviations, variance, inter-quartile range and outliers and correlation 

```python
print(f"Mean for storm speed is : {data["STORM_SPEED"].mean()}")
print(f"Median for storm speed is : {data["STORM_SPEED"].median()}")
print(f"Mode for storm speed is : {data["STORM_SPEED"].mode()}")
print(f"Maximum value for storm speed is : {data["STORM_SPEED"].max()}")
print(f"Minimum value for storm speed is : {data["STORM_SPEED"].min()}")
print(f"Standard deviation for storm speed is : {data["STORM_SPEED"].std()}")
print(f"Variance for storm speed is : {data["STORM_SPEED"].var()}")

q1 = data['STORM_SPEED'].quantile(0.25)
q3 = data['STORM_SPEED'].quantile(0.75)
iqr = q3-q1
print(f"Inter-quartile range is {iqr}")
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
storm_speed_outliers = data[(data['STORM_SPEED'] < lower_bound) & (data['STORM_SPEED'] > upper_bound)]
print(f"Outliers in Storm_speed values are : {storm_speed_outliers}")

data[['STORM_SPEED', 'STORM_DIR']].corr()
```

## Visualization of the data from the dataset we manipulated above using matplotlib.pyplot and seaborn

```python
data['STORM_SPEED'].hist(bins=20)
plt.title('DISTRIBUTION OF STORM SPEEDS')
plt.xlabel('STORM SPEED')
plt.ylabel('FREQUENCY')
plt.show()
```

```python
plt.scatter(data['STORM_SPEED'], data['STORM_DIR'])
plt.xlabel('STORM SPEED')
plt.ylabel('STORM DIRECTION')
plt.title('SCATTER PLOT SHOWING STORM SPEEDS VS DIRECTION OF THE STORM HEADING')
plt.show()
```
