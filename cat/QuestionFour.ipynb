# Question 4: ODE Simulation of Real System (20 marks)
You are simulating population growth using the differential equation `dy/dt = 0.5y`, where
y(0) = 100.


a) Use Euler’s Method to approximate the population size after 4 time steps with step size
h = 0.5. Show all computations. (8 Marks)

Euler’s formula:
- yn = yn-1 + h * f(yn-1)

Where f(y)=0.5y

- Step 0: 
y0 = 100 n=0, y0=100

- Step 1: 
y1 = 100 + 0.5 * 0.5 * 100 = 100 + 25 = 125

- Step 2: 
y2 = 125 + 0.5 * 0.5 * 125 = 156.25

- Step 3:
y3 = 156.25 * 0.5 * 0.5 * 156.25 = 156.25+ 39.0625 = 195.3125

- Step 4:
y4 = 195.3125 + 0.5 * 0.5 * 195.3125 = 195.3125 + 48.828125 = 244.140625

b) Write a Python function that generalizes the Euler’s method to solve any first-order
ODE given f(t, y), initial condition, step size, and number of steps. (6 Marks)

```python
def euler_method(f, t0, y0, h, steps):
    t, y = t0, y0
    results = [(t, y)]
    for _ in range(steps):
        y = y + h * f(t, y)
        t = t + h
        results.append((t, y))
    return results
```

c) Briefly explain two limitations of using Euler’s Method for real-world systems, and
suggest one alternative method. 

1) Euler's method of solving differential equations is only first-order accurate because things get complicated as orders of the differential equations increase.
2) Euler's method can be wildly inaccurate for rapidly changing systems(dynamic) in that it can produce wildly incorrect results or diverge.
Alternative: Runge-Kutta's 4th order(RK4) -> is more accurate and relatively simple to implement.
