## Question 3: Data Visualization and Interpretation (20 marks)
Given a file data.csv with two columns: Time (s) and Temperature (°C), your task is to
analyze a heat transfer process.

a) Write a Python script that:
- reads the data using pandas,
- plots a line graph of temperature vs time,
- and adds labels, title, and grid. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp_data = pd.read_csv('./temps.csv')
temp_data.columns = temp_data.columns.str.strip().str.replace(" ", "_").str.replace("°C", "C")
temp_data.head()
```

```python
plt.figure(figsize=(10,8))
plt.title("A Line Graph Showing a Relation of Temperature over Time")
plt.xlabel("Time(s)")
plt.ylabel("Temperature(°C)")
plt.plot(temp_data['Time(s)'], temp_data['Temperature_(C)'], marker='.')
```

b) Based on your graph, explain any two observations you can make about the heating
process (e.g., steady state, rate of change). 

1) The temperature steadily decreases over time, from the highest point which was near 30 degrees celcius to below 10 degrees celcius. This indicates cooling of the object displaying such temperatures.
2) The rate of cooling of the object displaying such temperatures is fairly fast, given it cools 22 degrees in just under 2 minutes. 

c) Describe a situation in engineering where such visualization can be used to influence
decision-making. 

- A situation:  an engineer is testing a prototype for a high performance CPU or an embedded device. The engineer wants to monitor the thermals of the CPU, so sensors are placed on the cpu and the value of temperature of the cpu over time is recorded from the values of the sensors. A graph can then be drawn using these values as a visual representation of the temperature of the CPU over time.
- Visualization helps to make the decision in that:
  1) Early spikes in the graph can indicate spikes in temperature which might indicate poor heat dissipation from the CPU, allowing the     engineer to perform changes accordingly, maybe add better cooling or thermal pads.
  2) Steady temperature rise, signified by a linearly rising line graph may signify inadequate cooling, meaning weak airflow or not enough cool air to cool the CPU.
  3) If the line graph balances out, it may mean the temperature is steady meaning that there is adequate airflow, and the system is effectively cooled.
